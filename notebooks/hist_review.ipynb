```python
import pandas as pd
import yfinance as yf

# Load the data from Excel
df = pd.read_excel("/Users/MilesChild/Desktop/Student Value Fund/pm/svf_code/data/purchases_history.xlsx")

# Create a date range for the required period
date_range = pd.date_range(start="2019-01-01", end="2023-09-10", freq="D")

# Initialize an empty dataframe for the portfolio
portfolio_df = pd.DataFrame({"Date": date_range})
portfolio_df.set_index("Date", inplace=True)

# Iterate over each asset in the Excel file
for index, row in df.iterrows():
    asset = row['asset']
    purchase_date = row['actual_purchase_date']
    qty = row['purchase_qty']
    
    # Fetch historical data using yfinance
    hist_data = yf.download(asset, start="2019-01-01", end="2023-09-10")
    
    # Only interested in the Close prices
    hist_data = hist_data[["Close"]]
    
    # Rename the column to the asset's name
    hist_data.rename(columns={"Close": asset}, inplace=True)
    
    # If this asset's data doesn't exist in the portfolio_df, add it
    if asset not in portfolio_df.columns:
        portfolio_df = portfolio_df.merge(hist_data, how="left", left_index=True, right_index=True)
        portfolio_df[asset].fillna(0, inplace=True)
    
    # Update the asset quantities based on purchase date
    portfolio_df.loc[purchase_date:, asset] += qty

# Calculate portfolio value for each day (price * quantity)
for asset in df['asset'].unique():
    hist_data = yf.download(asset, start="2019-01-01", end="2023-09-10")[["Close"]]
    portfolio_df[f"{asset}_value"] = portfolio_df[asset] * hist_data["Close"]

# Now, portfolio_df contains the quantities of all assets for each day
# and the corresponding value for each asset.
print(portfolio_df)
```

```python
import pandas as pd
import yfinance as yf

# Load the data from Excel
df_purchases = pd.read_excel("/Users/MilesChild/Desktop/Student Value Fund/pm/svf_code/data/purchases_history.xlsx")
df_sales = pd.read_excel("/Users/MilesChild/Desktop/Student Value Fund/pm/svf_code/data/sales_history.xlsx")

# Create a date range for the required period
date_range = pd.date_range(start="2019-01-01", end="2023-09-10", freq="D")

# Initialize an empty dataframe for the portfolio quantities and values
portfolio_qty_df = pd.DataFrame({"Date": date_range})
portfolio_qty_df.set_index("Date", inplace=True)

portfolio_value_df = portfolio_qty_df.copy()

# Process both purchase and sales dataframes in a loop
for df, action in [(df_purchases, 'buy'), (df_sales, 'sell')]:
    for index, row in df.iterrows():
        asset = row['asset']
        
        # Determine the correct date column based on the dataframe
        date_column = 'actual_purchase_date' if action == 'buy' else 'actual_sale_date'
        date = row[date_column]
        
        qty = row['purchase_qty'] if action == 'buy' else -row['sale_qty']

        # Fetch historical data using yfinance if not already fetched
        if asset not in portfolio_qty_df.columns:
            hist_data = yf.download(asset, start="2019-01-01", end="2023-09-10")
            portfolio_qty_df[asset] = 0.0
            portfolio_qty_df.loc[:date, asset] = qty  # Adjust quantities based on action date
            portfolio_value_df[asset] = portfolio_qty_df[asset] * hist_data["Close"]
        else:
            portfolio_qty_df.loc[date:, asset] += qty  # Adjust quantities based on action date
            hist_data = yf.download(asset, start="2019-01-01", end="2023-09-10")
            portfolio_value_df[asset] = portfolio_qty_df[asset] * hist_data["Close"]

print(portfolio_qty_df)
print(portfolio_value_df)

```

```python
# portfolio = quantities.copy()
# for t in quantities.keys():
#     asset_price_df = yf.download(t, start=start_date, end=end_date)
#     if len(asset_price_df) == 0:
#         continue
#     else:
#         asset_price = asset_price_df['Close']
    
#     for date in quantities[t].keys():
#         if date in asset_price.index:
#             portfolio[t][date] *= asset_price[date]
#         else:
#             continue

# portfolio = quantities.copy()

# for t in quantities.keys():
#     asset_price_df = yf.download(t, start=start_date, end=end_date)
    
#     # Continue if there's no data for this ticker
#     if len(asset_price_df) == 0:
#         continue
    
#     asset_price = asset_price_df['Close']
    
#     for date in quantities[t].keys():
#         # Convert the date from quantities to the same datetime format as asset_price
#         standardized_date = pd.to_datetime(date)
        
#         if standardized_date in asset_price.index:
#             portfolio[t][date] *= asset_price[standardized_date]
#         else:
#             continue

portfolio = quantities.copy()

for t in quantities.keys():
    asset_price_df = yf.download(t, start=start_date, end=end_date)
    
    # Continue if there's no data for this ticker
    if len(asset_price_df) == 0:
        continue
    
    asset_price = asset_price_df['Close']
    
    # Variable to track the last known price
    last_known_price = None
    
    for date in quantities[t].keys():
        # Convert the date from quantities to the same datetime format as asset_price
        standardized_date = pd.to_datetime(date)
        
        if standardized_date in asset_price.index:
            last_known_price = asset_price[standardized_date]
        else:
            # If date isn't found in asset_price, use the last known price
            if last_known_price is None:
                continue
        
        # Multiply by the current or last known price
        portfolio[t][date] *= last_known_price
        
        # Ensure the value is non-negative
        if portfolio[t][date] < 0:
            portfolio[t][date] = 0
```

```python
"""
We will now construct a portfolio dataframe with the index being a date range from the start_date to the end_date and there being two columns per ticker in tickers: <ticker>_qty and <ticker_price>.
"""

# create a date range where the dates are dt.date objects
date_range = pd.date_range(start_date, end_date).to_pydatetime().tolist()
date_range = [date.date() for date in date_range]
portfolio_df = pd.DataFrame(index=date_range)
for ticker in tickers:
    portfolio_df[ticker + '_qty'] = 0
    portfolio_df[ticker + '_price'] = 0
portfolio_df.head(10)
```

```python
# for each purchase in the purchases_df, add that purchase to the <ticker>_qty column in the portfolio_df and forward-fill the qty.
for index, row in purchases_df.iterrows():
    asset_column = row['asset'] + '_qty'
    
    # Update the quantity for the specified asset on the purchase date
    portfolio_df.at[row['actual_purchase_date'], asset_column] = row['purchase_qty']
    
    # Forward-fill the quantity for this asset
    portfolio_df[asset_column] = portfolio_df[asset_column].ffill()

# # Each time there is a purchase, I want to make the qty for that purchase to be the quantity in all future dates.
# portfolio_df = portfolio_df.fillna(method='ffill')
portfolio_df.tail()
```
